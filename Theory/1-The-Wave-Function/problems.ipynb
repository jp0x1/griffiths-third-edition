## Problem 1.2

#### (a) Find the standard deviation of the distribution in Example 1.2.

We define $\langle x \rangle =  \int_{-\infty}^{\infty} x\rho(x) \,dx$ and the standard deviation as $\sigma = \sqrt{\langle x^2 \rangle-\langle x \rangle^2}$

We know from Example 1.2 that the probability density $\rho(x) = 1/(2\sqrt{hx})$

We can then solve for $\langle x \rangle = \int_{0}^{h} x(\frac{1}{\sqrt{hx}}) \,dx = \frac{h}{3}$ 

We also know that $\langle x^2 \rangle = \int_{-\infty}^{\infty} x^2\rho(x) \,dx$

so $\langle x^2 \rangle = \int_{0}^{h} x^2\frac{1}{\sqrt{hx}} \,dx$, and therefore $\langle x^2 \rangle = \frac{h^2}{5}$

We know that standard deviation is $\sigma = \sqrt{\langle x^2 \rangle-\langle x \rangle^2} = \sqrt{\frac{h^2}{5} - \frac{h^2}{9}} = \frac{2h}{3\sqrt{5}}$

## Problem 1.3
Consider the gaussian distribution

$$ \rho(x) = A{\rm e}^{-\lambda(x-a)^2}$$

where $A$, $a$, and $\lambda$ are positive real constants.  (The necessary integrals are inside the back cover.)

#### (a) Use Equation 1.16 to determine $A$

We can use $\int_{-\infty}^{\infty} \rho(x)  \,dx = 1$ to normalize the function and solve for $A$

$\int_{-\infty}^{\infty} A{\rm e}^{-\lambda(x-a)^2}  \,dx = 1$

We know that there is a special integral,

$$ \int_{-\infty}^{\infty} {\rm e}^{-A(x)^2}  \,dx = \sqrt{\frac{\pi}{A}}$$

We can redefine our integral as 

$\int_{-\infty}^{\infty} A{\rm e}^{-\lambda(x-a)^2}  \,dx = 1$

$A \int_{-\infty}^{\infty} {\rm e}^{-\lambda(x-a)^2}  \,dx = 1$

$A \int_{-\infty}^{\infty} {\rm e}^{-\lambda(x-a)^2}  \,dx = 1$, 

Using u-substitution, we define $u$ as $(x-a)$, and can use the special integral to get our equation.

$A \int_{-\infty}^{\infty} {\rm e}^{-\lambda(u)^2}  \,du = 1$

$A \sqrt{\frac{\pi}{\lambda}} = 1$

$A = \sqrt{\frac{\lambda}{\pi}}$

#### (b) Find $\langle x \rangle$, $\langle x^2 \rangle$, and $\sigma$

$\langle x \rangle = \int_{-\infty}^{\infty} x\rho(x)  \,dx$

We defined $u = (x-a)$ so let's define $x = u + a$ with both $x$ and $\rho(x)$

$\langle x \rangle =\sqrt{\frac{\lambda}{\pi}} \int_{-\infty}^{\infty} (u+a) {\rm e}^{-\lambda(u)^2} \,du$

$\langle x \rangle =\sqrt{\frac{\lambda}{\pi}} \int_{-\infty}^{\infty} (u) {\rm e}^{-\lambda(u)^2} + (a) {\rm e}^{-\lambda(u)^2} \,du$

$\langle x \rangle =\sqrt{\frac{\lambda}{\pi}} [\int_{-\infty}^{\infty} (u) {\rm e}^{-\lambda(u)^2} \,du +  \int_{-\infty}^{\infty} (a) {\rm e}^{-\lambda(u)^2} \,du]$

We know that since $(u) {\rm e}^{-\lambda(u)^2}$ is an odd function with symmetry, $\int_{-\infty}^{\infty} (u) {\rm e}^{-\lambda(u)^2} \,du = 0$

We use the same special integral trick and get: 

$$\langle x \rangle =\sqrt{\frac{\lambda}{\pi}} [0 + a\sqrt{\frac{\pi}{\lambda}}] = a$$

$\langle x^2 \rangle = \int_{-\infty}^{\infty} x^2\rho(x)  \,dx$

We defined $u = (x-a)$ so let's define $x = u + a$ with both $x$ and $\rho(x)$

$\langle x^2 \rangle =\sqrt{\frac{\lambda}{\pi}} \int_{-\infty}^{\infty} (u+a)^2 {\rm e}^{-\lambda(u)^2} \,du$

$\langle x^2 \rangle =\sqrt{\frac{\lambda}{\pi}} \int_{-\infty}^{\infty} (u^2+2ua+a^2) {\rm e}^{-\lambda(u)^2} \,du$

We know that for $\int_{-\infty}^{\infty} (2ua) {\rm e}^{-\lambda(u)^2} \,du$ = 0 due to symmetry and being an odd function.

$\langle x^2 \rangle =\sqrt{\frac{\lambda}{\pi}} \int_{-\infty}^{\infty} (u^2+a^2) {\rm e}^{-\lambda(u)^2} \,du$

Using special [integrals](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTG55tszaegzB6WEVpz3fz9nKaa2ypw4-_Cp2bI9hs5S2hgutJuFzbWbLYqyY8QGyyCdQ&usqp=CAU), we find $\int_{-\infty}^{\infty} (u^2) {\rm e}^{-\lambda(u)^2} \,du = \frac{1}{2}\frac{\pi}{\lambda^3} $

$\langle x^2 \rangle =\sqrt{\frac{\lambda}{\pi}} (\frac{1}{2}\sqrt{\frac{\pi}{\lambda^3}} + a^2\sqrt{\frac{\pi}{\lambda}}) = (\frac{1}{2\lambda}+a^2)$

We find standard deviation through

$$ \sigma = \sqrt{\langle x^2 \rangle - \langle x \rangle^2}$$

$ \sigma = \sqrt{(\frac{1}{2\lambda}+a^2)-(a)^2}$

$ \sigma = \sqrt{(\frac{1}{2\lambda})} = \frac{1}{\sqrt{2\lambda}}$
